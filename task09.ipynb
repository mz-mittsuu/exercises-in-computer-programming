#### ã€Aèª²é¡Œã€‘ä»¥ä¸‹ã®ä»•æ§˜ã‚’æº€ãŸã™ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã›ã‚ˆ
* ä¹±æ•°ã®å€‹æ•° `numbers` ã€ä¹±æ•°ã®æœ€å°å€¤ `min_value` ã€ä¹±æ•°ã®æœ€å¤§å€¤ `max_value` ã‚’å…¥åŠ›
* å…¥åŠ›ã•ã‚ŒãŸå€¤ã«åŸºã¥ãã€æŒ‡å®šå€‹æ•°ã®æ•´æ•°ä¹±æ•°ã‚’ãƒªã‚¹ãƒˆã¨ã—ã¦ç”Ÿæˆã—è¡¨ç¤º
* å…ƒã®ä¹±æ•°ãƒªã‚¹ãƒˆã‚’ã‚³ãƒ”ãƒ¼ã—ã¦ã€ãƒãƒ–ãƒ«ã‚½ãƒ¼ãƒˆã‚’ç”¨ã„ã¦æ˜‡é †ã«ä¸¦ã¹æ›¿ãˆè¡¨ç¤º
* åŒæ§˜ã«ã€å…ƒã®ä¹±æ•°ãƒªã‚¹ãƒˆã‚’ã‚³ãƒ”ãƒ¼ã—ã¦ã€ãƒãƒ–ãƒ«ã‚½ãƒ¼ãƒˆã‚’ç”¨ã„ã¦é™é †ã«ä¸¦ã¹æ›¿ãˆè¡¨
ç¤º
* æ¤œè¨¼ã®ç‚ºã« `sorted` é–¢æ•°ã®ä½¿ç”¨ã¯å¯

```python
import random

# --- ãƒ¦ãƒ¼ã‚¶ãƒ¼å…¥åŠ› ---
num_values = int(input("ä¹±æ•°ã®å€‹æ•°ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„: "))
min_value = int(input("ä¹±æ•°ã®æœ€å°å€¤ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„: "))
max_value = int(input("ä¹±æ•°ã®æœ€å¤§å€¤ã‚’å…¥åŠ›ã—ã¦ä¸‹ã•ã„ã€‚: "))

# --- ä¹±æ•°ç”Ÿæˆ ---
random_values = [random.randint(min_value, max_value) for _ in range(num_values)]
print("ç”Ÿæˆã—ãŸä¹±æ•°:", random_values)


# ======================================================
# çµ„ã¿è¾¼ã¿ã‚½ãƒ¼ãƒˆ
# ======================================================
sorted_asc_builtin = sorted(random_values)
sorted_desc_builtin = sorted(random_values, reverse=True)


# ======================================================
# ãƒãƒ–ãƒ«ã‚½ãƒ¼ãƒˆï¼ˆæ˜‡é †ï¼‰ï¼‹æ¯”è¼ƒ/äº¤æ›ã‚«ã‚¦ãƒ³ãƒˆ
# ======================================================
bubble_sorted_asc = random_values[:]
length = len(bubble_sorted_asc)

asc_compare = 0
asc_swap = 0

for i in range(length - 1):
    for j in range(length - 1 - i):

        asc_compare += 1  # æ¯”è¼ƒå›æ•°ã‚«ã‚¦ãƒ³ãƒˆ

        if bubble_sorted_asc[j] > bubble_sorted_asc[j + 1]:
            # äº¤æ›
            bubble_sorted_asc[j], bubble_sorted_asc[j + 1] = bubble_sorted_asc[j + 1], bubble_sorted_asc[j]
            asc_swap += 1


# ======================================================
# ãƒãƒ–ãƒ«ã‚½ãƒ¼ãƒˆï¼ˆé™é †ï¼‰ï¼‹æ¯”è¼ƒ/äº¤æ›ã‚«ã‚¦ãƒ³ãƒˆï¼‹
# ======================================================
bubble_sorted_desc = random_values[:]
length = len(bubble_sorted_desc)

desc_compare = 0
desc_swap = 0

for i in range(length - 1):
    for j in range(length - 1 - i):

        desc_compare += 1  # æ¯”è¼ƒå›æ•°

        if bubble_sorted_desc[j] < bubble_sorted_desc[j + 1]:
            # äº¤æ›
            bubble_sorted_desc[j], bubble_sorted_desc[j + 1] = bubble_sorted_desc[j + 1], bubble_sorted_desc[j]
            desc_swap += 1


# ======================================================
# çµæœè¡¨ç¤º
# ======================================================
print("\n--- çµ„ã¿è¾¼ã¿ sorted() ã«ã‚ˆã‚‹æ˜‡é † ---")
print(sorted_asc_builtin)

print("\n--- ãƒãƒ–ãƒ«ã‚½ãƒ¼ãƒˆã«ã‚ˆã‚‹æ˜‡é †ï¼ˆçµæœï¼‰ ---")
print(bubble_sorted_asc)
print(f"æ¯”è¼ƒå›æ•°: {asc_compare}, äº¤æ›å›æ•°: {asc_swap}")

print("\n--- çµ„ã¿è¾¼ã¿ sorted() ã«ã‚ˆã‚‹é™é † ---")
print(sorted_desc_builtin)

print("\n--- ãƒãƒ–ãƒ«ã‚½ãƒ¼ãƒˆã«ã‚ˆã‚‹é™é †ï¼ˆçµæœï¼‰ ---")
print(bubble_sorted_desc)
print(f"æ¯”è¼ƒå›æ•°: {desc_compare}, äº¤æ›å›æ•°: {desc_swap}")
```

```python
# èª•ç”ŸçŸ³ãƒªã‚¹ãƒˆã‚’ä½œæˆ
stones = [
    "Garnets",
    "Amethyst",
    "Bloodstone:Aquamarine",
    "Diamond",
    "Emerald:Jade",
    "Moonstone",
    "Ruby",
    "Peridot:Sardonyx",
    "Sapphire",
    "Opal:Tourmaline",
    "Topaz",
    "Turquoise:Lapis lazuli",
]

try:
    # ãƒ¦ãƒ¼ã‚¶ãƒ¼ã«èª•ç”Ÿæœˆã‚’å…¥åŠ›ã•ã›ã‚‹
    mont = int(input("ã‚ãªãŸã®èª•ç”Ÿæœˆã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ (1-12): "))

    # å…¥åŠ›ãƒã‚§ãƒƒã‚¯
    if 1 <= mont <= 12:
        print(f"ã‚ãªãŸã®èª•ç”ŸçŸ³ã¯ {stones[mont - 1]} ã§ã™ã€‚")
    else:
        print("å…¥åŠ›ãƒŸã‚¹: 1ã‹ã‚‰12ã®æ•´æ•°ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚")

except ValueError:
    print("å…¥åŠ›ãƒŸã‚¹: æ•´æ•°ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚")
```

#### ã€Cèª²é¡Œã€‘ä»¥ä¸‹ã®ä»•æ§˜ã‚’æº€ãŸã™ãƒ—ãƒ­ã‚°ãƒ©ãƒ ã‚’ä½œæˆã›ã‚ˆ
* ãƒ¦ãƒ¼ã‚¶ãƒ¼ã‹ã‚‰è¡Œæ•° `N` ã€åˆ—æ•° `M` ã€ä¹±æ•°ã®æœ€å°å€¤ã€æœ€å¤§å€¤ã‚’å…¥åŠ›
* ãã®å€¤ã‚’ä½¿ã£ã¦ `N Ã— M` ã®2æ¬¡å…ƒãƒªã‚¹ãƒˆï¼ˆè¡Œåˆ—ï¼‰ã‚’ç”Ÿæˆãƒ»è¡¨ç¤º
* å„è¡Œã®åˆè¨ˆå€¤ã‚’è¨ˆç®—ã—ã¦è¡¨ç¤ºã—ã€åˆè¨ˆå€¤ãŒå¤§ãã„é †ã«è¡Œã”ã¨ã«ã‚½ãƒ¼ãƒˆã—ã¦2æ¬¡å…ƒãƒªã‚¹ãƒˆã‚’è¡¨ç¤º

```python
import random

# --- 1. ãƒ¦ãƒ¼ã‚¶ãƒ¼ã‹ã‚‰ã®å…¥åŠ› ---
try:
    rows = int(input("è¡Œæ•° N ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„: "))
    cols = int(input("åˆ—æ•° M ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„: "))
    min_val = int(input("ä¹±æ•°ã®æœ€å°å€¤ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„: "))
    max_val = int(input("ä¹±æ•°ã®æœ€å¤§å€¤ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„: "))
except ValueError:
    print("ã‚¨ãƒ©ãƒ¼: æ•´æ•°ã‚’å…¥åŠ›ã—ã¦ãã ã•ã„ã€‚")
    exit()

# --- 2. N x M ã®2æ¬¡å…ƒãƒªã‚¹ãƒˆã‚’ãƒªã‚¹ãƒˆå†…åŒ…è¡¨è¨˜ã§ç”Ÿæˆ ---
matrix = [[random.randint(min_val, max_val) for _ in range(cols)] for _ in range(rows)]

print("\n--- ç”Ÿæˆã•ã‚ŒãŸå…ƒã®è¡Œåˆ— (N x M) ---")
for row in matrix:
    print(row)

# ------------------------------------------------------------------

## 3. ğŸ“Š å…ƒã®åˆè¨ˆå€¤ã®è¡¨ç¤ºã¨ã‚½ãƒ¼ãƒˆå‡¦ç†

# å„è¡Œã®åˆè¨ˆå€¤ã‚’è¨ˆç®—
row_sums = [sum(row) for row in matrix]

# å…ƒã®è¡Œåˆ—ã®åˆè¨ˆå€¤ã‚’è¡¨ç¤º (ä»Šå›è¿½åŠ ã—ãŸéƒ¨åˆ†)
print(f"\nå…ƒã®è¡Œåˆ—ã®å„è¡Œã®åˆè¨ˆå€¤: {row_sums}")

# sorted() é–¢æ•°ã‚’ä½¿ç”¨ã—ã¦ã€matrix (è¡Œã®ãƒªã‚¹ãƒˆ) ã‚’ã‚½ãƒ¼ãƒˆã—ã¾ã™ã€‚
# key=sum: ã‚½ãƒ¼ãƒˆã®åŸºæº–ã¨ã—ã¦ã€å„è¡Œ(ãƒªã‚¹ãƒˆ)ã®åˆè¨ˆå€¤ã‚’ä½¿ç”¨ã€‚
# reverse=True: åˆè¨ˆå€¤ã®å¤§ãã„é †ï¼ˆé™é †ï¼‰ã«ä¸¦ã¹æ›¿ãˆã‚‹ã€‚
sorted_matrix_by_sum = sorted(matrix, key=sum, reverse=True)

# ------------------------------------------------------------------

## 4. ğŸ–¨ï¸ ã‚½ãƒ¼ãƒˆçµæœã®è¡¨ç¤º

# ã‚½ãƒ¼ãƒˆå¾Œã®å„è¡Œã®åˆè¨ˆå€¤ã‚’ç¢ºèªç”¨ã¨ã—ã¦è¨ˆç®—
sorted_sums = [sum(row) for row in sorted_matrix_by_sum]

print("\n--- ã‚½ãƒ¼ãƒˆçµæœï¼ˆåˆè¨ˆå€¤ã®é™é †ï¼‰ ---")
print(f"ã‚½ãƒ¼ãƒˆå¾Œã®åˆè¨ˆå€¤ã®ä¸¦ã³: {sorted_sums}")
print("ã‚½ãƒ¼ãƒˆã•ã‚ŒãŸè¡Œåˆ—:")
for row in sorted_matrix_by_sum:
    print(row)
```
