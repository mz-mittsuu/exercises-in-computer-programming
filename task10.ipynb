#### 【B課題】以下の仕様を満たすプログラムを作成せよ

* ユーザーに文字列を入力させる
* 入力文字中のアルファベットのみを対象に出現回数をカウント
* 出現回数は辞書型に格納
* 辞書のキーをアルファベット順に並べて出力
* 各文字の出現回数を「文字：個数個」の形式で表示
* 大文字・小文字は区別する
* 検証のために `from collections import Counter` で `Counter` ：リストの各要素の数え上げを使用は可


```python
print("文字列を入力してください:")
input_string = input()

# 文字の出現回数を記録する辞書（Dictionary）
# キーが文字、値が個数になります
counts = {}

# 文字列を1文字ずつ走査
for char in input_string:
    # アルファベット（A-Z, a-z）のみを対象とする
    if char.isalpha():# アルファベットかどうかを確認
        #if ('A' <= char <= 'Z') or ('a' <= char <= 'z'): # アルファベットの範囲を確認不要
            # 辞書に文字がなければ0で初期化し、あれば+1する
            counts[char] = counts.get(char, 0) + 1 # 文字の出現回数をカウント
            """# 以下は同じ処理をif文で書いた場合
            if char in counts:
               counts[char] += 1   # すでにキーがあれば +1
            else:
               counts[char] = 1    # キーがなければ初期値1で追加
            """

print("\n各アルファベット文字の含まれている個数:")
"""
from collections import Counter
filtered_chars = [c for c in input_string if ('A' <= c <= 'Z') or ('a' <= c <= 'z')]
# 一気にカウント
counts = Counter(filtered_chars)
"""
# 辞書をキー（文字）でソートして表示
for char in sorted(counts.keys()):
    print(f"{char}: {counts[char]}個")# 文字とその出現回数を表示
```

#### 【C課題】以下の仕様を満たすプログラムを作成せよ

* `[9,1,3,4,6,7,8,]` ほぼソート<br>
最大要素：先頭, 1回のパスで1ずつのみ後方に移動<br>
→ ソート完了：比較21回, 交換6回を要する
* 奇数パスでは最小要素を先頭側に移動<br>
偶数パスでは最大要素を末尾側に移動<br>
パスの走査方法を交互に変更<br>
→ 双方向バブルソート（シェーカーソート）
* 課題9Aをベースに同じ配列に対して単純バブルソート, シェーカーソートでソートして2つのソート方法での比較・交換回数を表示せよ<br>
<span style="color: blue; ">交換回数は変わらないはず</span>
* <span style="color: red; ">昇順のみで良い<br>
特定配列に関する問題ではない<br>
検証用配列：`[9,1,3,4,6,7,8]` に対して行いシェーカーソートは比較11回, 交換6回となるはず</span>


```python
import random

# --- ユーザー入力 ---
print("検証用配列9,1,3,4,6,7,8を使いますか？(y/n): ")
use_test_array = input().lower() == 'y'

if use_test_array:
    random_values = [9, 1, 3, 4, 6, 7, 8]
else:
    print("乱数の個数を入力してください: ")
    num_values = int(input())
    print("乱数の最小値を入力してください: ")
    min_value = int(input())
    print("乱数の最大値を入力してください: ")
    max_value = int(input())
    random_values = [random.randint(min_value, max_value) for _ in range(num_values)]

print("元のリスト:", random_values)

# ======================================================
# 組み込みソート
# ======================================================
sorted_asc_builtin = sorted(random_values)

# ======================================================
# シェーカーソート（昇順）＋比較/交換カウント
# ======================================================
shaker_sorted_asc = random_values[:]
n = len(shaker_sorted_asc)
shaker_compare = 0
shaker_swap = 0

left = 0
right = n - 1
swapped = True

while swapped:
    swapped = False
    # 左から右へ
    for i in range(left, right):
        shaker_compare += 1
        if shaker_sorted_asc[i] > shaker_sorted_asc[i + 1]:
            shaker_sorted_asc[i], shaker_sorted_asc[i + 1] = shaker_sorted_asc[i + 1], shaker_sorted_asc[i]
            shaker_swap += 1
            swapped = True
    if not swapped:
        break
    right -= 1
    # 右から左へ
    for i in range(right, left, -1):
        shaker_compare += 1
        if shaker_sorted_asc[i - 1] > shaker_sorted_asc[i]:
            shaker_sorted_asc[i - 1], shaker_sorted_asc[i] = shaker_sorted_asc[i], shaker_sorted_asc[i - 1]
            shaker_swap += 1
            swapped = True
    left += 1

# ======================================================
# バブルソート（昇順）＋比較/交換カウント
# ======================================================
bubble_sorted_asc = random_values[:]
bubble_compare = 0
bubble_swap = 0
for i in range(n - 1):
    for j in range(n - 1 - i):
        bubble_compare += 1
        if bubble_sorted_asc[j] > bubble_sorted_asc[j + 1]:
            bubble_sorted_asc[j], bubble_sorted_asc[j + 1] = bubble_sorted_asc[j + 1], bubble_sorted_asc[j]
            bubble_swap += 1

# ======================================================
# 結果表示
# ======================================================
print("\n--- 組み込み sorted() による昇順 ---")
print(sorted_asc_builtin)

print("\n--- シェーカーソートによる昇順（結果） ---")
print(shaker_sorted_asc)
print(f"比較回数: {shaker_compare}, 交換回数: {shaker_swap}")

print("\n--- バブルソートによる昇順（結果） ---")
print(bubble_sorted_asc)
print(f"比較回数: {bubble_compare}, 交換回数: {bubble_swap}")
```
