#### 【A課題】整数 `n` を入力して `n!`（n の階乗）を計算するプログラムを作成せよ

* マイナスの整数は計算不能とすること
* `0!` の場合は `1` を出力すること
* 変数名として `number` と `factorial` の2つを用いる事
* `math.factorial` のみの使用は不可（※検証に記載はOK）

```python
number = int(input("整数を入力してください: "))

if number < 0:
    print("マイナスの整数の階乗は計算できません。")
elif number == 0:
    print("0! = 1")
else:
    factorial = 1
    # i の代わりに k を使用 (数学的な慣例)
    for k in range(1, number + 1):
        factorial *= k
    print(f"{number}! = {factorial}")

"""
# mathモジュールを使う
import math

# number を使用
if number >= 0:
    print(f"【math版】{number}! = {math.factorial(number)}")
else:
    print("【math版】マイナスの整数の階乗は計算できません。")
"""
```

#### 【B課題】以下の動作を実現するプログラムを作成せよ

* データ件数が100件未満の1学年分の学生の体重[kg]と身長[cm]のデータがある
* このデータを学生１人分(体重・身長)ずつ入力し、データを全て打ち込み終わったら体重または身長を0と入力し、入力を終了
* 入力数と体重・身長・ボディマス指数BMIの平均値を求めよ

$$\mathrm{BMI} = \frac{\text{体重 [kg]}}{(\text{身長 [m]})^2}$$

* 日本肥満学会基準で痩せすぎ、適正体重、肥満が何人いるか表示せよ

| BMI条件 | 判定区分 |
|:---:|:---:|
| $\mathrm{BMI} < 18.5$ | 痩せすぎ |
| $18.5 \leq \mathrm{BMI} < 25.0$ | 適正体重 |
| $\mathrm{BMI} \geq 25.0$ | 肥満 |

```python
# BMI計算プログラム（最大100件）

print("学生の体重(kg)と身長(cm)を入力してください。")
print("体重または身長に0を入力すると終了します。\n")

# 初期化
sum_w = 0.0       # 体重の合計
sum_h = 0.0       # 身長の合計
sum_bmi = 0.0     # BMIの合計
count = 0         # 入力人数カウント

thin = 0          # 痩せすぎ人数
normal = 0        # 適正体重人数
fat = 0           # 肥満人数

# 最大100件の入力をfor文で処理
for i in range(100):
    print(f"{i+1}人目のデータ")
    weight = float(input("体重(kg): "))
    height = float(input("身長(cm): "))

    # 入力終了判定
    if weight == 0 or height == 0:
        print("入力を終了します。\n")
        break

    # BMI計算
    height_m = height / 100
    bmi = weight / (height_m ** 2)

    # 結果表示
    print(f"BMI = {bmi:.1f}")

    # 合計・カウント更新
    sum_w += weight
    sum_h += height
    sum_bmi += bmi
    count += 1

    # 体格分類
    if bmi < 18.5:
        thin += 1
    elif bmi < 25:
        normal += 1
    else:
        fat += 1

# 平均値の出力（1人以上入力された場合）
if count > 0:
    print("\n--- 結果 ---")
    print(f"入力人数: {count}")
    print(f"平均体重: {sum_w / count:.1f} kg")
    print(f"平均身長: {sum_h / count:.1f} cm")
    print(f"平均BMI: {sum_bmi / count:.1f}")
    print(f"痩せすぎ: {thin}人, 適正体重: {normal}人, 肥満: {fat}人")
else:
    print("データが入力されませんでした。")
```

#### 【C課題】次の仕様に従って、 `*` を使って直角三角形を描くプログラムを作成せよ。

1. ユーザーから短辺の長さ（整数）を入力で受け取る
2. 次に、描く位置を「左下」「左上」「右下」「右上」「中央」「逆」のいずれかで入力する
3. 長さ、入力された位置に応じて対応する形の三角形を出力する
4. 指定以外の位置が入力された場合は、エラーメッセージを表示する

```python
# 直角三角形を描くプログラム
# match文を使用

n = int(input("短辺の長さを入力してください: ")) # 短辺の長さを受け取る
pos = input("描く位置を入力してください（左下, 左上, 右上, 右下, 中央, 逆）: ") # 位置の指定を受け取る

# match文を使用して条件分岐
"""
二重ループ

外側ループ：行数の制御

内側ループ：1行の中の文字（空白または*）の個数を制御

最後に print() で改行して次の行に移動
"""

match pos:
    case "左下":
        for i in range(1, n + 1):
            for j in range(i):
                print("*", end="")
            print()

    case "左上":
        for i in range(n, 0, -1):
            for j in range(i):
                print("*", end="")
            print()

    case "右下":
        for i in range(1, n + 1):
            for j in range(n - i):
                print(" ", end="")
            for j in range(i):
                print("*", end="")
            print()

    case "右上":
        for i in range(n, 0, -1):
            for j in range(n - i):
                print(" ", end="")
            for j in range(i):
                print("*", end="")
            print()

    case "中央":
        for i in range(1, n + 1):
            for j in range(n - i):
                print(" ", end="")
            for j in range(2 * i - 1):
                print("*", end="")
            print()

    case "逆":
        for i in range(n, 0, -1):
            for j in range(n - i):
                print(" ", end="")
            for j in range(2 * i - 1):
                print("*", end="")
            print()

    case _:
        print("指定が正しくありません。左下, 左上, 右下, 右上, 中央, 逆 のいずれかを入力してください。")

# 改良版: 文字列の繰り返しを使用して簡潔に記述
"""
match pos:
    case "左下":
        for i in range(1, n + 1):
            print("*" * i)

    case "左上":
        for i in range(n, 0, -1):
            print("*" * i)

    case "右下":
        for i in range(1, n + 1):
            print(" " * (n - i) + "*" * i)

    case "右上":
        for i in range(n, 0, -1):
            print(" " * (n - i) + "*" * i)

    case "中央":
        for i in range(1, n + 1):
            print(" " * (n - i) + "*" * (2 * i - 1))

    case "逆":
        for i in range(n, 0, -1):
            print(" " * (n - i) + "*" * (2 * i - 1))

    case _:
        print("指定が正しくありません。左下, 左上, 右下, 右上, 中央, 逆 のいずれかを入力してください。")
"""
```
