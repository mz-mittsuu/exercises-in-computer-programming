#### 【A課題】以下の仕様を満たすプログラムを作成せよ

* `main()` 関数を定義し、ユーザから整数を1つ入力
* 整数の桁数を計算する関数 `count_digits()` を呼び出し、「入力された数字は○桁です。」の形式で結果を出力
* 整数の桁数を計算する関数 `count_digits()` は以下を満たすこと
    * 引数 `number` の桁数を整数として返す
    * `0` の場合は `1` を返す
    * 負の数が入力された場合はその絶対値で桁数を求める

```python
def count_digits(number):
    """整数の桁数を返す関数"""
    #return len(str(abs(number))) # 別解

    # 0 の場合は 1 桁
    if number == 0:
        return 1

    # 負の数は正にする
    if number < 0:
        number = -number

    digit_count = 0

    # 桁数を計算
    while number > 0:
        number //= 10
        digit_count += 1

    return digit_count


def main():
    number = int(input("整数を入力してください。\n"))

    digit_count = count_digits(number)

    print(f"入力された数字は{digit_count}桁です。")

# 実行部分
main()
```

#### 【B課題】以下の仕様のビンゴカード生成プログラムを作成せよ

* ユーザが入力した整数 `N (3 ~ 7)` に基づき、 `N × N` のビンゴカードを作成し表示する
* ユーザー定義関数を用いること
* 各マスには `1 ~ 75` の整数をランダムで重複しないように配置
* ` N` が奇数の場合、中央（行と列の中間位置）は文字列 `FREE` とする
* `random.shuffle` は使用可


```python
import random

def generate_bingo_card(n):
    """1～75の数字から n×n のビンゴカードを生成して2次元リストで返す"""
    numbers = list(range(1, 76))
    random.shuffle(numbers)

    # n×n のマス分取り出し
    selected = numbers[:n * n]

    # 2次元リストに変換
    card = []
    index = 0
    for _ in range(n):
        row = []
        for _ in range(n):
            row.append(selected[index])
            index += 1
        card.append(row)

    # 奇数のとき真ん中を FREE
    if n % 2 == 1:
        mid = n // 2
        card[mid][mid] = "FREE"

    return card


def print_bingo_card(card):
    """ビンゴカードをきれいに表示する"""
    print("\n=== Bingo Card ===")
    for row in card:
        for item in row:
            print(f"{str(item):>4}", end=" ")
        print()
    print()


def main():
    print("=== N×N ビンゴカード生成プログラム ===")
    n = int(input("ビンゴカードのサイズ N を入力してください (3～7): "))

    if n < 3 or n > 7:
        print("N は 3 以上 7 以下で入力してください。")
        return

    bingo_card = generate_bingo_card(n)
    print_bingo_card(bingo_card)


main()
```

#### 【C課題】以下の仕様を満たすプログラムを作成せよ

* 課題9Aをベースに同じ配列に対して単純バブルソート・単純挿入ソート(シャトルソート)でソートして2つのソート方法で比較・交換回数を表示する<br>
<span style="color: blue; ">交換回数は変わらないはず</span>
* <span style="color: red; ">昇順のみで良い<br>
特定配列に関する問題ではない<br>
検証用リスト：`[6,4,1,7,3,9,8]` に対して行いシャトルソートは比較11回, 交換7回となるはず</span>

```python
import random

# ----------------------------
# 乱数リスト生成
# ----------------------------
def generate_list():
    use_test_array = input("検証用配列[6,4,1,7,3,9,8]を使いますか？(y/n): ").lower() == 'y'

    if use_test_array:
        return [6, 4, 1, 7, 3, 9, 8]

    num_values = int(input("乱数の個数を入力してください: "))
    min_value = int(input("乱数の最小値を入力してください: "))
    max_value = int(input("乱数の最大値を入力してください: "))

    return [random.randint(min_value, max_value) for _ in range(num_values)]


# ----------------------------
# シャトルソート（Shuttle Sort）
# ----------------------------
def shuttle_sort(values):
    arr = values[:]  # コピーして操作
    compare = 0
    swap = 0
    n = len(arr)

    for i in range(1, n):
        j = i
        while j > 0:
            compare += 1
            if arr[j - 1] > arr[j]:
                arr[j - 1], arr[j] = arr[j], arr[j - 1]
                swap += 1
                j -= 1
            else:
                break

    return arr, compare, swap


# ----------------------------
# シェーカーソート（Shaker Sort）
# ----------------------------
def shaker_sort(values):
    arr = values[:]
    n = len(arr)
    compare = 0
    swap = 0
    left = 0
    right = n - 1
    swapped = True

    while swapped:
        swapped = False
        # 左→右
        for i in range(left, right):
            compare += 1
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                swap += 1
                swapped = True

        if not swapped:
            break
        right -= 1

        # 右→左
        for i in range(right, left, -1):
            compare += 1
            if arr[i - 1] > arr[i]:
                arr[i - 1], arr[i] = arr[i], arr[i - 1]
                swap += 1
                swapped = True

        left += 1

    return arr, compare, swap


# ----------------------------
# バブルソート
# ----------------------------
def bubble_sort(values):
    arr = values[:]
    n = len(arr)
    compare = 0
    swap = 0

    for i in range(n - 1):
        for j in range(n - 1 - i):
            compare += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swap += 1

    return arr, compare, swap


# ----------------------------
# 結果表示
# ----------------------------
def print_result(title, sorted_list, compare, swap):
    print(f"\n--- {title} ---")
    print(sorted_list)
    print(f"比較回数: {compare}, 交換回数: {swap}")


# ----------------------------
# メイン処理
# ----------------------------
def main():
    random_values = generate_list()
    print("元のリスト:", random_values)

    # 組み込みソート
    builtin_sorted = sorted(random_values)
    print("\n--- 組み込み sorted() による昇順 ---")
    print(builtin_sorted)

    # シャトルソート
    shuttle_list, shuttle_compare, shuttle_swap = shuttle_sort(random_values)
    print_result("シャトルソート（昇順）", shuttle_list, shuttle_compare, shuttle_swap)

    # シェーカーソート
    shaker_list, shaker_compare, shaker_swap = shaker_sort(random_values)
    print_result("シェーカーソート（昇順）", shaker_list, shaker_compare, shaker_swap)

    # バブルソート
    bubble_list, bubble_compare, bubble_swap = bubble_sort(random_values)
    print_result("バブルソート（昇順）", bubble_list, bubble_compare, bubble_swap)


# ----------------------------
# 実行
# ----------------------------
main()
```
