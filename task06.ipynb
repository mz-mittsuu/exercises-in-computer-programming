#### 【B課題】以下の動作を満たすプログラムを作成せよ

1. 乱数生成
    - 0 ~ 99 の乱数 `target_number`を生成する
    - 生成した乱数は表示しない
2. 入力と判定
    - 生成した乱数とキーボードから入力した整数 `guess` が **等しくなれば** 終了する
    - 等しくなければ再度キーボードから入力した数字と比較する
3. 試行回数の表示
    - 何回でヒットしたか表示する
4. ヒントの表示
    - 等しくないときヒントを表示
    - 大きい、小さい、xx以上離れている、など
5. 独自機能
    - ランダムな誤ヒント：<br>
        ランダムな確率で嘘のヒントを表示
    - 得点システム：<br>
        正解に到達するまでの試行回数や、何回か間違えると乱数をリセットし、その回数に応じてスコアを減点するなど独自機能があってもよい

```python
# 〇 1: 0〜99の乱数target_numberを生成
# 〇 2: target_numberとguessが等しくなるまでループ
# 〇 3: 何回でヒットしたか表示 (attempts)
# 〇 4: ヒントを表示（大きい、小さい、近い、遠い）
# 〇 5: 独自機能（ランダムな嘘ヒント、スコア機能）

import random

# --- 1. 乱数の準備 ---
# 0から99までの整数 target_number (正解の数) を生成
target_number = random.randint(0, 99) # rからtarget_numberに変更

# 提出時は以下をコメントアウトします
# print(f"(デバッグ用) 乱数 target_number は {target_number} です")

# --- 変数の初期化 ---
attempts = 0  # 試行回数 (countからattemptsに変更)
score = 1000  # スコア (Requirement 5)

print("0〜99の乱数 target_number が発生しました。当ててください")

# --- 2. メインループ (while文) ---
while True:
    # ユーザーからの入力を受け取る
    guess_str = input("0〜99の整数 guess を入力して下さい: ")

    # --- 入力値のチェック (発展的) ---
    try:
        guess = int(guess_str)
        if not (0 <= guess <= 99):
            print("エラー: 0〜99の数を入力してください。")
            continue
    except ValueError:
        print("エラー: 整数を入力してください。")
        continue

    # --- 3. 回数とスコアの更新 ---
    attempts += 1 # 回数を更新
    # 1回試行するごとに50点減点 (最低点は0)
    score = max(0, score - 50)

    # --- 判定 ---
    if guess == target_number:
        # 2. 等しくなれば終了
        print(f"\n--- 🎉 正解です！ ---")
        print(f"{attempts} 回で当たりました！") # 3. ヒット回数を表示
        print(f"あなたのスコア: {score} 点") # 5. スコア表示
        break  # whileループを脱出
    else:
        # --- 4 & 5. ヒントの表示 ---

        # 5. 独自機能: 20% (1/5) の確率で嘘のヒントを表示
        if random.randint(1, 5) == 1:
            # 嘘ヒント
            if random.randint(0, 1) == 0:
                print("ヒント: ...もっと大きいです (嘘のヒント)")
            else:
                print("ヒント: ...もっと小さいです (嘘のヒント)")
        else:
            # 4. 正しいヒント
            hint = ""
            if guess < target_number:
                hint = "もっと大きいです"
            else:
                hint = "もっと小さいです"

            # 4. 独自機能: 距離に応じたヒントを追加
            diff = abs(target_number - guess) # 差の絶対値
            if diff >= 20:
                hint += " (かなり離れています)"
            elif diff <= 5:
                hint += " (🎯 とても近いです！)"

            print(f"ヒント: {hint}")

    # スコアが0点になったらゲームオーバー (独自機能)
    if score == 0:
        print("\n--- 💀 ゲームオーバー ---")
        print(f"スコアが0になりました。正解は {target_number} でした。")
        break
```

#### 【C課題】以下の仕様のひし形を表示する **While** 文を用いたプログラムを作成せよ

* ユーザーからひし形の中心の幅(奇数)を入力する
* 入力された値が正の整数であり、かつ奇数でない場合は再入力

```python
# ------------------------------------
# 1. 無限ループで奇数の幅を入力させる
# ------------------------------------
width = 0
# while True で無限ループを開始
while True:
    # ユーザーからひし形の中心の幅を入力してもらう
    input_str = input("ひし形の中心の幅（奇数の正の整数）を入力してください: ")
    # int()で整数に変換（非整数を入力するとここでエラーになります）
    width = int(input_str)

    # 1. 正の整数であること (width > 0)
    # 2. 奇数であること (width % 2 != 0)
    if width > 0 and width % 2 != 0:
        break  # 条件を満たしたらループを抜ける
    elif width <= 0:
        print("正の整数を入力してください。")
    else:
        print("奇数 (例: 1, 3, 5, 7...) を入力してください。")

# ------------------------------------
# 2. ひし形の表示ロジック
# ------------------------------------

# 「大きさ」（中心から端までの最大幅、つまり半分の高さ）を計算
# 例: 幅 5 -> size 3, 幅 7 -> size 4
size = (width + 1) // 2

# --- 上半分と中央の行 ---
row = 1
# row が size 以下の間、ループを繰り返す
while row <= size:
    # スペースの数: size - row
    space_count = size - row
    i = 0
    # スペースを出力
    while i < space_count:
        print(" ", end="")
        i += 1

    # アスタリスクの数: 2 * row - 1
    star_count = 2 * row - 1
    j = 0
    # アスタリスクを出力
    while j < star_count:
        print("*", end="")
        j += 1

    print()  # 行の終わりに改行
    row += 1

# --- 下半分 (中央の次の行から) ---
row = size - 1
# row が 1 以上の間、ループを繰り返す
while row >= 1:
    # スペースの数: size - row
    space_count = size - row
    i = 0
    # スペースを出力
    while i < space_count:
        print(" ", end="")
        i += 1

    # アスタリスクの数: 2 * row - 1
    star_count = 2 * row - 1
    j = 0
    # アスタリスクを出力
    while j < star_count:
        print("*", end="")
        j += 1

    print()  # 行の終わりに改行
    row -= 1
```
