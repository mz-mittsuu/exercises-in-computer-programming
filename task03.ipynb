#### 【A課題】整数を1つ入力し、その整数が偶数であるか奇数であるかを判定して表示するプログラムを作成せよ

```python
# 偶数・奇数を判定するプログラム

# 整数を入力
n = int(input("整数を入力してください: "))

# 偶数・奇数を判定
if n % 2 == 0:
    print(f"{n} は偶数です。")
else:
    print(f"{n} は奇数です。")
```

#### 【B課題】以下を満たすプログラムを作成せよ

* キーボードから身長[cm]、体重[kg]を入力・表示
* ボディマス指数(BMI)を計算し、少数第2位まで表示

$$\mathrm{BMI} = \frac{\text{体重 [kg]}}{(\text{身長 [m]})^2}$$

* 日本肥満学会基準で痩せすぎ、適正体重、肥満が何人いるか表示せよ

| BMI条件 | 判定区分 |
|:---:|:---:|
| $\mathrm{BMI} < 18.5$ | 痩せすぎ |
| $18.5 \leq \mathrm{BMI} < 25.0$ | 適正体重 |
| $\mathrm{BMI} \geq 25.0$ | 肥満 |

```python
# 身長(cm)と体重(kg)を入力してBMIを計算し、判定するプログラム

# 入力
height = float(input("身長を[cm]単位で入力してください: "))
weight = float(input("体重を[kg]単位で入力してください: "))

# cm → m に変換
height_m = height / 100

# BMI計算
bmi = weight / (height_m ** 2)

# BMI値を小数第2位まで表示
print(f"身長{height} [cm]、体重{weight} [kg]のBMIは{bmi:.2f}")

# 判定
if bmi < 18.5: # 18.5未満
    print("痩せすぎ (Underweight)")
elif bmi < 25: # 18.5以上25未満
    print("適正体重 (Healthy weight)")
else: # 25以上
    print("肥満 (Overweight)")
```
