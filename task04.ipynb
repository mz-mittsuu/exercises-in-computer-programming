#### 【A課題】シリコンウエハーの購入枚数を入力し、マイナスや、25の倍数でない場合売れませんと表示し、25の倍数であれば合計支払い金額・消費税を計算し、画面に表示させるプログラムを作成せよ

* シリコンウエハーはウエハー容器に25枚入れられクリーンパッケージされている為、25枚単位でしか販売できない
* 4インチウエハー1枚あたり消費税(10%)込みで3000円で販売中で、購入枚数が100枚以上で1枚2700円、500枚以上で1枚2500円の大口購入割引有り

```python
# シリコンウェーハの購入金額を計算するプログラム

# 説明文の表示
print("シリコンウェーハは25枚単位で販売されている")
print("4インチシリコンウェーハ1枚の税込価格は以下の通りである")
print("100枚未満: 3000円")
print("100枚以上、500枚未満: 2700円")
print("500枚以上: 2500円")

# 枚数の入力
n = int(input("4インチのシリコンウェーハを何枚購入したいですか？: "))

# 条件判定：25の倍数でない、または負の値なら販売不可
if n % 25 != 0 or n < 0:
    print(f"{n}枚では売れません。")
else:
    # 枚数に応じた税込単価を決定
    if n >= 500:
        p = n * 2500
    elif n >= 100:
        p = n * 2700
    else:
        p = n * 3000

    # 消費税（内税方式なので税込価格から税分を逆算）
    tax = int(p * (0.10 / 1.10))

    # 結果の出力
    print(f"合計金額は {p}円で、消費税は {tax}円です。")
```

#### 【B課題】以下の映画館のチケット料金を計算・表示するプログラムを作成せよ

* 入力1:会員かどうかの入力は英語または日本語で行えるようにせよ
    会員種別（`member` または `会員` ／ `general` または `非会員`）
* 入力2:年齢
* 不正な会員種別が入力された場合は「エラー: 不正な会員種別です。」と表示する

|区分|60歳以上|18歳未満|18歳以上60歳未満|
|---|---|---|---|
|`member` / `会員`|1000円|1200円|1500円|
|`general` / `非会員`|1200円|1500円|2000円|

```python
# 映画料金計算プログラム
member_type = input("会員種別を入力してください （会員 member/非会員 general): ")# 会員種別は文字列で入力
age = int(input("年齢を入力してください: "))# 年齢は整数値で入力

# 料金を格納する変数を初期化
price = 0

# match文を使用した料金計算
match member_type:
    case 'member' | '会員':
        # 会員料金の判定
        match age:
            case a if a >= 60:
                price = 1000
            case a if a < 18:
                price = 1200
            case _: # 18歳以上60歳未満
                price = 1500
        """ # 会員料金の判定（if文）
        if age >= 60:
            price = 1000
        elif age < 18:
            price = 1200
        else:  # 18歳以上60歳未満
            price = 1500
         """
    case 'general'| '非会員':
        # 非会員料金の判定
        match age:
            case a if a >= 60:
                price = 1200
            case a if a < 18:
                price = 1500
            case _: # 18歳以上60歳未満
                price = 2000
        """ # 非会員料金の判定（if文）
        if age >= 60:
            price = 1200
        elif age < 18:
            price = 1500
        else:  # 18歳以上60歳未満
            price = 2000
        """

    case _:
        price = -1 # 不正な会員種別
        print("エラー: 不正な会員種別です。")

# 結果の出力
if price > 0:
    print(f"会員種別: {member_type}, 年齢: {age}歳")
    print(f"映画料金は {price} 円です。")
```
