#### 【B課題】キーボードから身長[cm]を入力し、以下の要件を満たすプログラムを作成せよ

* 入力された身長を表示する
* 以下の式で適正体重[kg]を計算する<br>
$$適正体重[\mathrm{kg}] = 22 \times (身長[\mathrm{m}]) \times (身長[\mathrm{m}])$$
* 計算された適正体重を小数第2位まで表示する

```python
# 身長（cm）の入力を受け付ける
height_cm = float(input("身長をセンチメートル[cm]で入力してください: "))
print(f"入力された身長: {height_cm} [cm]")
# 身長をcmからmに変換
height_m = height_cm / 100

# 適正体重を計算
ideal_weight = 22 * (height_m ** 2)

# 結果を小数第1位まで表示
print(f"適正体重: {ideal_weight:.2f} [kg]")
```

#### 【C課題】食事会を行った際の支払いに関するプログラムを作成せよ

* キーボードから食事会の総額, 参加人数, 割引クーポン額を入力
* クーポン適用後の支払総額、1人当たりの支払金額(整数)、割り切れない場合の余りの金額を表示

```python
# 飲み会の総額、参加人数、割引クーポン額を入力してもらう
total, people, coupon = input(
        "食事会の総額 参加人数 割引クーポン額 をスペース区切りで入力してください（例: 25000 6 3000）: "
    ).split()
# 型変換
total = int(total)      #食事会の総額
people = int(people)    # 人数
coupon = int(coupon)    # 割引クーポン
# クーポンを引いた金額
net_total = total - coupon

# 一人当たりの金額（整数除算）
per_person = net_total // people

# 割り切れない余り
remainder = net_total % people

# 結果を表示（f-string使用）
print(f"割引後の総額: {net_total} 円")
print(f"一人当たりの支払い額: {per_person} 円")
print(f"余りの金額: {remainder} 円")
```
